# 기초통계량(평균, 중앙값, IQR, 표준편차 등)

## 1) 중심측도를 나타내는 값(평균, 중앙값, 최빈값)

```python
df.shape # dataframe의 행렬을 알 수 있음.
```

```python
# 평균값 구하기
mpg_mean = df['mpg'].mean()
print(mpg_mean)
```

```python
# 중앙값 구하기
mpg_median = df['mpg'].median()
print(mpg_median)
```

```python
# 최빈값 구하기
mpg_mode = df['cyl'].mode()
print(mpg_mode)
```

```python
print(mpg_mode[0]) # 값만 가져오고 싶을때는 이렇게 가져와야 함.
```

```python
df['cyl'].value_counts()
```

## 2) 산포도를 나타내는 값(뿐산, 표준편가, IQR, 범위(최대-최소) 등)

```python
# 분산
mpg_var = df['mpg'].var()
```

```python
print(mpg_var)
```

```python
# 표준편차
mpg_std = df['mpg'].std()
print(mpg_std)
```

```python
# IQR
Q1 = df['mpg'].quantile(0.25)
print(Q1)
```

```python
Q3 = df['mpg'].quantile(0.75)
print(Q3)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
Q2 = df['mpg'].quantile(0.5)
print(Q2) # median값이랑 같음.
```

```python
#	범위(Range)	=	최대값	-	최소값
mpg_max	=	df['mpg'].max()
print(mpg_max)
```

```python
mpg_min	=	df['mpg'].min()
print(mpg_min)
```

```python
 mpg_range	=	mpg_max	-	mpg_min
 print(mpg_range)
```

## 3) 분포의 비대칭도

```python
 #	왜도
mpg_skew	=	df['mpg'].skew()
 print(mpg_skew)
 #	첨도
mpg_kurt	=	df['mpg'].kurt()
 print(mpg_kurt)
```

## 4) 기타(합계, 절대값, 데이터 수 등)

```python
#	합계
mpg_sum	=	df['mpg'].sum()
print(mpg_sum)

 #	절대값
IQR2	=	Q1-Q3
print(IQR2)
print(abs(IQR2))


 #	데이터	수
len(df['mpg'])
```
